```python
# Which state has highest production?
production_by_state = df.groupby('State_Name')['Production'].sum()
state_with_highest_production = production_by_state.idxmax()
highest_production_value = production_by_state.max()
state_with_highest_production
```

```python
df.groupby(["State_Name"]).agg({"Production" : "sum"}).plot(kind = "bar")
```

```python
# which crop has highest area?
ax = sns.barplot(data = df.head(10),  x= "Crop", y = "Area")
ax.set_xticklabels(ax.get_xticklabels(),rotation =90)
plt.show()
```

```python
# which season has highest production and how much?
max_season =df.loc[df['Production'].idxmax(), 'Season']
max_production =df['Production'].max()
print(f"The season with the highest production is {max_season} with a production of {max_production}")
```

```python
# which year has highest production?
ax = sns.barplot(data=df,  x= "Crop_Year", y = "Production",color="brown")
ax.set_xticklabels(ax.get_xticklabels(),rotation =90)
plt.show()
```
